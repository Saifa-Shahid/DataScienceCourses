# Feature Extraction

```python
data.dtypes
```

```python
data.columns
```

```python
lag_features=['High','Low','Volume','Turnover','Trades']

window=3
window=7
```

```python
for feature in lag_features:
    data[feature+'rolling_mean_3']=data[feature].rolling(window).mean()
    data[feature+'rolling_mean_7']=data[feature].rolling(window).mean()
```

```python
for feature in lag_features:
    data[feature+'rolling_std_3']=data[feature].rolling(window).std()
    data[feature+'rolling_std_7']=data[feature].rolling(window).std()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
inde_features=['Highrolling_mean_3', 'Highrolling_mean_7',
       'Lowrolling_mean_3', 'Lowrolling_mean_7', 'Volumerolling_mean_3',
       'Volumerolling_mean_7', 'Turnoverrolling_mean_3',
       'Turnoverrolling_mean_7', 'Tradesrolling_mean_3',
       'Tradesrolling_mean_7', 'Highrolling_std_3', 'Highrolling_std_7',
       'Lowrolling_std_3', 'Lowrolling_std_7', 'Volumerolling_std_3',
       'Volumerolling_std_7', 'Turnoverrolling_std_3', 'Turnoverrolling_std_7',
       'Tradesrolling_std_3', 'Tradesrolling_std_7']
```

```python
training_data=data[0:1800]
test_data=data[1800:]
```

```python
training_data
```

```python
test_data
```

```python
!pip install pmdarima
```

```python
from pmdarima import auto_arima
```

```python
model=auto_arima(y=training_data['VWAP'],exogenous=training_data[inde_features],trace=True)
```

```python
model.fit(training_data['VWAP'],training_data[inde_features])
```

```python
forecast=model.predict(n_periods=len(test_data),exogenous=test_data[inde_features])
```

```python
test_data['Forecast_ARIMA']=forecast
```

```python
test_data[['VWAP','Forecast_ARIMA']].plot(figsize=(14,7))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.sqrt(mean_squared_error(test_data['VWAP'],test_data['Forecast_ARIMA']))
```

```python
mean_absolute_error(test_data['VWAP'],test_data['Forecast_ARIMA'])
```
